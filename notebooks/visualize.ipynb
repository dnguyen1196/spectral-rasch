```python
def plot_top_k_accuracy(res, k_ind):
    K_array = res['K_array']
    K = K_array[k_ind]
    dataset = res["dataset"]
    n_trials = len(res['trial_seeds'])
    p_array = res['p_array']
    
    top_k_ase = np.array(res['top_k_ase'])
    top_k_mmle = np.array(res['top_k_mmle'])
    
    ase_accuracy = np.mean(top_k_ase[:, :, k_ind], 0)
    mmle_accuracy = np.mean(top_k_mmle[:, :, k_ind], 0)
    
    fig, ax = plt.subplots()
    ax.plot(p_array, ase_accuracy, label="ASE")
    ax.plot(p_array, mmle_accuracy, label="MMLE")
    ax.legend()
    ax.set_xlabel("Portion of training data")
    ax.set_ylabel(f"accuracy")
    ax.set_title(f"{dataset} top-{K} accuracy")
    fig.tight_layout()
    return fig, ax


def plot_heldout_auc(res):
    dataset = res["dataset"]
    n_trials = len(res['trial_seeds'])
    p_array = res['p_array']
    
    auc_ase = np.mean(np.array(res['auc_ase']), 0)
    auc_mmle = np.mean(np.array(res['auc_mmle']), 0)

    fig, ax = plt.subplots()
    ax.plot(p_array, auc_ase, label="Spectral")
    ax.plot(p_array, auc_mmle, label="MMLE")
    ax.legend()
    ax.set_xlabel("Portion of training data")
    ax.set_ylabel(f"AUC")
    ax.set_title(f"{dataset} test AUC")
    fig.tight_layout()
    return fig, ax


def plot_heldout_loglik(res):
    dataset = res["dataset"]
    n_trials = len(res['trial_seeds'])
    p_array = res['p_array']
    
    loglik_ase = np.mean(np.array(res['loglik_ase']), 0)
    loglik_mmle = np.mean(np.array(res['loglik_mmle']), 0)

    fig, ax = plt.subplots()
    ax.plot(p_array, loglik_ase, label="Spectral")
    ax.plot(p_array, loglik_mmle, label="MMLE")
    ax.legend()
    ax.set_xlabel("Portion of training data")
    ax.set_ylabel(f"log likelihood")
    ax.set_title(f"{dataset} test log likelihood")
    fig.tight_layout()
    return fig, ax


def plot_time(res):
    dataset = res["dataset"]
    n_trials = len(res['trial_seeds'])
    p_array = res['p_array']
    
    time_ase = np.array(res['time_ase'])
    top_k_mmle = np.array(res['top_k_mmle'])
    
    ase_accuracy = np.mean(top_k_ase[:, :, k_ind], 0)
    mmle_accuracy = np.mean(top_k_mmle[:, :, k_ind], 0)
    
    fig, ax = plt.subplots()
    ax.plot(p_array, ase_accuracy, label="Spectral")
    ax.plot(p_array, mmle_accuracy, label="MMLE")
    ax.legend()
    ax.set_xlabel("Portion of training data")
    ax.set_ylabel(f"accuracy")
    ax.set_title(f"{dataset} top-{K} accuracy")
    fig.tight_layout()
    return fig, ax
    
    
    
```
